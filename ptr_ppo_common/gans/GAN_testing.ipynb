```python
import torch
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time 

from gans import ptr_gan
```

```python
cuda = True if torch.cuda.is_available() else False

if cuda:
    print("CUDA is active")
else:
    print("CUDA is not active")
```

```python
import random
import math
dat = []
for i in range(400):
    x = random.uniform(-10,10)
    y = math.cos(x)
    dat.append((x,y))
cos_data = pd.DataFrame(dat,columns=["x","y"])
cos_data.head()

norm_cos_data = 2 * (cos_data-cos_data.min())/(cos_data.max()-cos_data.min()) -1
print(norm_cos_data.min())
print(norm_cos_data.max())
```

```python
plt.scatter(cos_data['x'],cos_data['y'])
```

```python
cos_tensor = torch.tensor(cos_data.values)
cos_tensor = cos_tensor.to(torch.float32)
dimensions = cos_tensor[0].shape[0]
print(cos_tensor)
if cuda:
    print("CUDA is active")
    cos_tensor = cos_tensor.cuda()


cos_gan = LTMGAN.LTMGAN(dimensions)
```

```python
start = time.time()
cos_gan.train(load_list=cos_tensor)
end = time.time()
```

```python
print('Total time training: ',(end-start)/60, ' minutes')
```

```python
plt.title("Gan Loss")
plt.plot(cos_gan.g_loss)
plt.plot(cos_gan.d_loss)
```

```python
Tensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor
```

```python
gen_x = []
gen_y = []
cos_gan.current_generator.eval()
for i in range(300):
    z = Variable(Tensor(np.random.normal(-1, 1, (1, cos_gan.latent_dim))))
    gen_ten = cos_gan.current_generator(z).detach()
    gen_li = gen_ten.tolist()
    x, y = gen_li[0]
    gen_x.append(x)
    gen_y.append(y)
    
```

```python
plt.title("Cos(x) = y")
plt.scatter(gen_x,gen_y)
plt.scatter(cos_data['x'],cos_data['y'])
print(max(gen_x))
print(min(gen_x)) 
```
